```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings

from scipy.stats.mstats import winsorize

warnings.filterwarnings("ignore")
```

```python
usa=pd.read_excel("veri_kumeleri/states_all.xls", encoding="utf8")
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.boxplot(usa["TOTAL_REVENUE"])

plt.subplot(1,2,2)
plt.boxplot(usa["TOTAL_EXPENDITURE"])

plt.show()
```

```python
plt.subplot(1,2,1)
plt.boxplot(usa["TOTAL_REVENUE"], whis=7)
#plt.title("Görüntülenme sayısının kutu grafiği (whis=20)", fontdict=baslik_font)

plt.subplot(1,2,2)
plt.boxplot(usa["TOTAL_EXPENDITURE"], whis=7)

plt.show()
```

```python
from scipy.stats import zscore

z_scores = zscore(usa["TOTAL_REVENUE"])
for threshold in range(1,10):
    print("Eşik değeri: {}".format(threshold))
    print("Aykırı değerlerin sayısı: {}".format(len((np.where(z_scores > threshold)[0]))))
    print('------')

z_scores = zscore(usa["TOTAL_EXPENDITURE"])
for threshold in range(1,10):
    print("Eşik değeri: {}".format(threshold))
    print("Aykırı değerlerin sayısı: {}".format(len((np.where(z_scores > threshold)[0]))))
    print('------')
```

```python
plt.figure(figsize = (15,5))

plt.subplot(1,2,1)
plt.boxplot(usa["TOTAL_REVENUE"])
plt.title("TOTAL_REVENUE")

plt.subplot(1,2,2)
plt.boxplot(np.log(usa["TOTAL_REVENUE"]))
plt.title("TOTAL_REVENUE (log dönüşümlü)")
plt.show()
```

```python
# 3
```

```python
usa["BUDGET_DEFICIT"] = usa["TOTAL_REVENUE"] - usa["TOTAL_EXPENDITURE"]
```

```python
z_scores = zscore(usa["BUDGET_DEFICIT"])
for threshold in range(1,10):
    print("Eşik değeri: {}".format(threshold))
    print("Aykırı değerlerin sayısı: {}".format(len((np.where(z_scores > threshold)[0]))))
    print('------')
```

```python
from scipy.stats.mstats import winsorize

winsorize_goruntulenme = winsorize(usa["BUDGET_DEFICIT"], (0, 0.10))

winsorize_goruntulenme
```

```python
plt.boxplot(winsorize_goruntulenme,whis=23)
plt.title("BUDGET_DEFICIT  kutu grafiği (whis=23)")
plt.show()
```

```python
usa["TOTAL_REVENUE_clean"] = np.log(usa["TOTAL_REVENUE"])
usa["TOTAL_EXPENDITURE_clean"] = np.log(usa["TOTAL_EXPENDITURE"])
usa["BUDGET_DEFICIT_clean"] = usa["TOTAL_REVENUE_clean"] - usa["TOTAL_EXPENDITURE_clean"]
```

```python
z_scores = zscore(usa["BUDGET_DEFICIT_clean"])
for threshold in range(1,10):
    print("Eşik değeri: {}".format(threshold))
    print("Aykırı değerlerin sayısı: {}".format(len((np.where(z_scores > threshold)[0]))))
    print('------')
```
