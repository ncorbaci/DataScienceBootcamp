```python
import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sns
from scipy import stats

standart_normal_dagilimli = np.random.normal(5, 10, 10000)
standart_normal_dagilimli.sort()
plt.figure(figsize=(8,4), dpi = 100)
plt.hist(standart_normal_dagilimli, bins=20, color = 'c')

plt.show()
```

```python
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }

plt.figure(figsize=(10,4), dpi = 200)
#nsample = 100
res = stats.probplot(standart_normal_dagilimli, plot=plt)
plt.title("Q-Q Çizgesi", fontdict = baslik_font)
plt.xlabel("Teorik Değerler", fontdict = eksen_font)
plt.ylabel("Örnek Veriler", fontdict = eksen_font)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('GOOG.csv')
pricelist=df["Close"]
pricelen=len(pricelist)
kazanc=[]

n=1

while (n < pricelen):
    kazanc.append((pricelist[n]-pricelist[n-1])/pricelist[n-1])
    n += 1
    
plt.hist(kazanc)
plt.show()

```

```python
normal_dagilimli_veri = np.random.normal(np.mean(kazanc),np.std(kazanc), len(kazanc))
normal_dagilimli_veri.sort()
kazanc.sort()

plt.figure(figsize=(8,5), dpi = 200)
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }
plt.title('Kazanç Veri Kümesi & Normal Dağılımlı Veri Kümesi', fontdict = baslik_font)
plt.xlabel('Kazanç Veri Kümesi', fontdict = eksen_font)
plt.ylabel('Normal Dağılımlı Veri Kümesi', fontdict = eksen_font)
plt.scatter(normal_dagilimli_veri, kazanc)
plt.plot([np.min(normal_dagilimli_veri), np.max(normal_dagilimli_veri)],
         [np.min(kazanc), np.max(kazanc)], c='red')
plt.show()

```
