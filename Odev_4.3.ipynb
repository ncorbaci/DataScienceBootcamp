```python
# YEAR kolonunun tüm değerleri 2017 olduğu için boş olan değerli de 2017 yapmak veri analizi için uygundur
# STATE kolonu PRIMARY_KEY alanın _ alt çizgiden sonraki değer ile aynı olduğu için PRIMARY_KEY e bakarak değer verilebilir
# Diğer kolonlar sürekli değişkene uyduğu için ortalama, medyan veya mod değeri ile doldurulabilir
usa["TOTAL_REVENUE"].fillna(usa["TOTAL_REVENUE"].mean(), inplace=True)
usa
```

```python
#3
#usa.fillna(usa.groupby("YEAR").mean(), inplace=True)
#usa["TOTAL_REVENUE"].fillna(usa["TOTAL_REVENUE"].mean(), inplace=True)
usa2 = pd.read_csv("veri_kumeleri/states_all.csv")
usa2['TOTAL_REVENUE'] = usa2.groupby(['YEAR'])['TOTAL_REVENUE'].transform(lambda x: x.fillna(x.mean()))
usa2
```
